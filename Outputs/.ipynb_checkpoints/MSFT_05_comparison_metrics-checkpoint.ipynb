```python
import matplotlib.pyplot as plt
import pandas as pd

# Strategy performance data
metrics_df = pd.DataFrame({
    'Buy & Hold': {
        'Total Return (%)': 176.53,
        'Annualized Volatility (%)': 30.51,
        'Sharpe Ratio': 0.82,
        'Max Drawdown (%)': -37.15
    },
    'Basic EMA (10/50)': {
        'Total Return (%)': 83.55,
        'Annualized Volatility (%)': 21.36,
        'Sharpe Ratio': 0.68,
        'Max Drawdown (%)': -25.99
    },
    'Triple EMA (10/50/200)': {
        'Total Return (%)': 71.99,
        'Annualized Volatility (%)': 19.81,
        'Sharpe Ratio': 0.65,
        'Max Drawdown (%)': -24.26
    },
    'EMA (10/50) + ATR SL': {
        'Total Return (%)': 109.52,
        'Annualized Volatility (%)': 27.78,
        'Sharpe Ratio': 0.68,
        'Max Drawdown (%)': -37.15
    },
    'Triple EMA + ATR SL': {
        'Total Return (%)': 137.40,
        'Annualized Volatility (%)': 26.62,
        'Sharpe Ratio': 0.79,
        'Max Drawdown (%)': -37.15
    },
})

# Transpose for plotting
metrics_df_T = metrics_df.T

# Strategy colors
strategy_colors = {
    'Buy & Hold': '#1f77b4',
    'Basic EMA (10/50)': '#ff7f0e',
    'Triple EMA (10/50/200)': '#2ca02c',
    'EMA (10/50) + ATR SL': '#d62728',
    'Triple EMA + ATR SL': '#9467bd'
}

# Define desired metric order
metric_order = [
    'Sharpe Ratio',
    'Total Return (%)',
    'Max Drawdown (%)',
    'Annualized Volatility (%)'
]

# Plot setup
fig, axs = plt.subplots(nrows=4, ncols=1, figsize=(10, 14))

for i, metric in enumerate(metric_order):
    values = metrics_df_T[metric].sort_values(ascending=True)  # Sort bars
    colors = [strategy_colors.get(strategy, '#333333') for strategy in values.index]

    axs[i].barh(values.index, values, color=colors)
    axs[i].set_title(metric, fontsize=13)
    axs[i].invert_yaxis()
    axs[i].grid(True, axis='x', linestyle='--', alpha=0.6)

    # Dynamic xlim padding
    min_val, max_val = values.min(), values.max()
    range_span = max_val - min_val
    axs[i].set_xlim(min_val - 0.05 * range_span, max_val + 0.1 * range_span)

    # Annotation
    for idx, val in enumerate(values):
        align = 'left' if val >= 0 else 'right'
        offset = range_span * 0.01 if val >= 0 else -range_span * 0.01
        axs[i].text(val + offset, idx, f'{val:.2f}', va='center', ha=align, fontsize=9)

# Final layout
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.savefig("strategy_comparison_metrics.png", dpi=300, bbox_inches='tight')
plt.show()
```
